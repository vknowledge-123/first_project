```python
import time
class Admin :
    def sign_up_manager(self):
        print("🌟 WELCOME TO ADMIN SITE! 🛍️")
        self.admin_dict = {}
        print("-------- Sign Up ---------")
        print("-----------------------------------")
        Admin_Id = input("Enter 👤Admin_ID:--- ")
        Passwd = input("Enter Admin 🔒Password :--- ")
        print("------------[ Sign In ]------------")
        print("-----------------------------------")
        self.admin_dict = {"Admin_ID" :Admin_Id ,"Password" : Passwd}
        print(self.admin_dict)
        print("✅ Admin account creation was successful !!")
        
    def login_manager(self):
        adminname = input("Enter AdminID: ")
        password = input("Enter Admin_Password: ")
        if self.admin_dict["Admin_ID"] == adminname and self.admin_dict["Password"] == password:
            print("Login successful! Welcome, Admin.")
        else:
            print("Invalid credentials. Please try again.")
            
    def start(self):
        self.sign_up_manager()
        self.login_manager()
        shop = Product_Manager()
        print("\n----Insert Product Inventory----")
        shop.shirts_section()
        shop.jeans_section()
        shop.list_details()
        sign = Authentication()
        sign.sign_up()
        sign.login()
        return shop

class Authentication:
    def sign_up(self):
        self.user_dict = {}
        print("\n🌟 WELCOME TO OUR SITE! 🛍️")
        print("***************************")
        print("--Sign In To Your Account--")
        print("👤 Username: _____")
        print("🔒 Password: _____")
        print("-------- Sign Up ---------")
        print("I’m new here, let me register")
        print("-----------------------------------")
        print("--Create Your Account--")
        print("-----------------------------------")
        user = input("Enter your 👤Username :--- ")
        passwd = input("Enter your 🔒Password :--- ")
        print("------------[ Sign In ]------------")
        self.user_dict = {"Username": user, "Password": passwd }
        print(self.user_dict)
        print("✅ Account created successful !!")
    
    def login(self):
        print("\n--------- Log In ---------")
        print("I already have an account, let me log in.")
        Username = input("Enter your 👤Username :--- ")
        if Username != self.user_dict["Username"]:
            print("❌ Invalid Username! Please enter a valid username.")
            self.login()
        else:
            attempts = 3
            while attempts > 0 :
                Password = input("enter ur 🔒Password :--- ")
                if self.user_dict["Password"] == Password:
                    print("✅ User Verified")
                    print("🔑 Login Successful !!")
                    break
                else:
                    attempts -= 1
                    if attempts == 0:
                        print("Too many failed attempts.\n")
                        print("Account blocked for 24 hours")
                        from datetime import datetime, timedelta
                        current_time = datetime.now()
                        print("Current Time:", current_time)
                        future_time = current_time + timedelta(hours=24)
                        print("Try Login Tomorrow at :", future_time)
                    print("Exiting.......")
                    print(f"⚠️Invalid Password. {attempts} attempts left..\n")

class Product_Manager:
        shirts = {}
        jeans = {}
        products = {}
    
        def shirts_section(self):
            print("---------------------------------")
            print(" Fill the details for Shirts : ")
            product_id = (input("Enter the product ID : "))
            name = input("Enter the name of the product: ")
            price = int(input("Enter the price of product :"))
            quantity = int(input("Enter the quantity of products : "))
            self.shirts = {"ID": product_id, "Name": name, "Price": price, "Quantity": quantity}
            print("Shirt added:", self.shirts)

        def jeans_section(self):
            print("--------------------------------")
            print(" Fill the details for Jeans : ")
            product_id = (input("Enter the product ID : "))
            name = input("Enter the name of the product : ")
            price = int(input("Enter the price of product :"))
            quantity = int(input("Enter the quantity of products : "))
            self.jeans = {"ID": product_id,"Name": name, "Price": price, "Quantity": quantity}
            print("Jeans added:", self.jeans)

        def list_details(self):
            print("\n---List of Product Inventory Details ---")
            self.products = {"Shirts" : self.shirts, "Jeans" : self.jeans}
            time.sleep(1)
            print(self.products) 
            
        def display_products(self):
            print("\nAvailable Products:")
            for product, details in self.products.items():
                print(f"{product} - Price: {details['Price']} | Stock: {details['Quantity']}")


class ShoppingSystem:
    def display_menu(self):
        print("\n-----🌟Welcome to our Shopping site !🌟-----")
        
        print("\n----Display Menu----")
        print("1️⃣ View Products")
        print("2️⃣ Remove add to cart")
        print("3️⃣ Update Stock")
        print("4️⃣ Payment")
        print("5️⃣ Exit")
    
    def select_option(self,c,stock,pay,shop):
        while True:
            self.display_menu()
            x=int(input("select option [1-5] :-  "))
            if x==1:
                self.view_products(c,shop)
            elif(x==2):
                c.remove_cart(shop)
            elif(x==3):
                stock.inventory(c,shop)
            elif(x==4):
                pay.get_total_bill(shop,c)
            elif(x==5):
                print("Exiting the shopping system.")
                print("✨ Stay connected & shop again soon! 🎁🛒")
                break
            else:
                print("Please Try again !")
                
    def view_products(self, c,shop):
        print("\n----Display All The Products----")
        print("👕 Shirts")
        print("👖 Jeans")
        print("👚 T-Shirts")
        print("🧥 Jackets")
        shop.display_products()
        while True:
            print("\n✨Welcome to our online store!")
            print("\nWould you like to add this product to your shopping cart? [y/n] ")
            opt = input("Enter :- [y/n]")
            if opt=="y":
                print("---Take a look at this amazing product.---")
                break
            else:
                print(" No Purchased product")
        while True:
            product=input("\nEnter the name of Product : ")
            if product in ["shirts", "jeans"]:
                c.add_cart_by_customer()
                print("Thanks for the Shopping !")
                z = input("Would you like to add one more product? [y/n] : ")
                if z != "y":
                    break
            else:
                print("Invalid product. Try again.")
                    
class Add_Cart:
    cart_dict = {}
        
    def add_cart_by_customer(self):
        print("\n-------Add to Cart------")
        print("✨ Start shopping now & let the magic begin! ✨")
        item = input("Enter the product name to add : ")
        quantity = int(input("Enter quantity to add : "))
        
        print("Searching for your details.....")
        self.cart_dict[item] = quantity
        print("Product added to the cart successful")
        print("🛒 Happy Shopping! 🛒")
        print("🛒 Item Added to Cart! ✅ :-- ", self.cart_dict)
    
    def remove_cart(self, shop):
        print("-----Remove item quantity form stock-----")
        for prod, qty in self.cart_dict.items():
            if prod == "shirts" and "Shirts" in shop.products:
                shop.products["Shirts"]["Quantity"] -= qty
                print("🗑️ Product quantity Removed from Cart! ✅")
                print("Shirts = ", shop.products["Shirts"])
            elif prod == "jeans" and "Jeans" in shop.products:
                shop.products["Jeans"]["Quantity"] -= qty
                print("Jeans = ", shop.products["Jeans"])
            else:
                print(f"Product '{prod}' not found in inventory.")

class Update_Stocks_By_Manager:
    def __init__(self,invent):
        self.shop = invent
        
    def inventory(self, c, shop):
        print("\n----Update the Stock----")
        quantity = int(input("Enter quantity to update the stock : "))
        for item, qty in c.cart_dict.items():
            if item == "shirts":
                shop.shirts["Quantity"] += quantity
                print("📉 Stock of Shirts Updated! 🔄:", shop.shirts)
            elif item == "jeans":
                shop.jeans["Quantity"] += quantity
                print("📉 Stock of Jeans Updated! 🔄:", shop.jeans)
            else:
                print("Invalid Stock or No Shirts in Inventory")
        
class Bill:
    total_bill = 0
    def get_total_bill(self, shop, c):
        print("\n----Calculating Total Bill----")
        print("\n---------- 🧾 Your Bill is Ready! 💰 ----------")
        print("📜Check your total and proceed to payment securely.\n")
        print("🛒 Product Purchased")
        print("-"*50)
        print(f"{'Item':<15}{'Price (₹)':<15}{'Quantity':<10}{'Total (₹)'}")
        print("-" * 50)
        
        for item, quantity in c.cart_dict.items():
            if item == "shirts":
                price = shop.shirts["Price"]
                total_price = price * quantity
                self.total_bill = self.total_bill + total_price
            elif(item == "jeans"):
                price = shop.jeans["Price"]
                total_price = price * quantity
                self.total_bill = self.total_bill + total_price
            print(f"{item.capitalize():<15}{price:<15}{quantity:<10}{total_price}")
        
        print("-" * 50)
        print(f"{'Total Bill:':<40} ₹{self.total_bill}")
        print("-" * 50)
        print(f"💰Your Total Bill is : ₹{self.total_bill}")
        print("Payment Successful!✅ ")
        print("🙏 YOUR ORDER IS CONFIRMED!")
        print("🛍️ Thank You for Shopping with Us! 🛍️")
        
        opt = input("Would u like to cancel the order [y/n] : ")
        if opt == "y":
            confirm = input("Please confirm: Do you really want to cancel? [y/n]: ")
            if confirm == "y":
                self.cancel_order(c)
            else:
                print("👍Your order is processed")
        
    def cancel_order(self,c):
        c.cart_dict = {}
        print("✅ Your order is cancelled ",c.cart_dict)
        print("💰 Your refund will be processed in 7 working days")
        
manager = Admin()
shop = manager.start()
cart = ShoppingSystem()
c = Add_Cart()
stock = Update_Stocks_By_Manager(shop)
pay = Bill()
cart.select_option(c,stock,pay,shop)
```
